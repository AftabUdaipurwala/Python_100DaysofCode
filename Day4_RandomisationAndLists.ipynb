# Random Module

```python
import random
c = random.randint(10,20)
print(c)
```

# Heads or Tails

```python
import random	 
	 
h_t= random.randint(0,1)
if h_t==1:
    print('Heads')
else:
    print('Tails')
```

# Lists

```python
fruits = ['apple','mango', 'cherry', 'orange']
```

```python
# access items in the list
fruits[0]
```

```python
# access by negative index
fruits[-1]
```

```python
# update list
fruits[3] = 'Oranges'
print(fruits  )
```

```python
# appended at the end
fruits.append('banana')
print(fruits)
```
