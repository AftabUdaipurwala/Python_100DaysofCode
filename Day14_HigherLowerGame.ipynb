# Functions

```python
import random


def randselector(list_original,first_value):
    list_new = list_original.copy()
    list_new.remove(first_value)
    return random.choice(list_new)

def value_switcher_a(data,compare_a):
    data_a = data.copy()
    compare_a = compare_a
    compare_b = randselector(data_a, compare_a)
    return data_a,compare_a,compare_b

```
