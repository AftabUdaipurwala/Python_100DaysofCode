# Mathematical operations

Rule is PEMDAS
1. Execute whatever is in parenthesis
2. Then do exponent
3. Then do multiplication
4. Then Do Division
5. Then do Addition
6. Last is subraction

since multiplication== division and Addition == subraction then calculation goes from left to right

```python
print(3*3+3/3+3**3-3 )
```

```python
print(3*(3+3)/3-3)
```

```python
# 🚨 Don't change the code below 👇
age = input("What is your current age?")
# 🚨 Don't change the code above 👆

#Write your code below this line 👇
year_left = 90 - int(age)
months = 12* year_left
weeks = 52*year_left
days = 365*year_left

print(f'You have {days} days, {weeks} weeks, and {months} months left.')
```

# Final Excercise : Tip Calculator at the restaurant

```python
total_bill = float(input('What was the total Bill?\n$'))
percent_tip = float(input('What percentage tip would you like to give? 10, 12 or 15 \n'))
people_split = int(input('How many people to split the bill? \n'))


result = round((total_bill+(total_bill*(percent_tip/100)))/people_split,2)

message = f'Each person should pay : ${result}'
print(message)
```
