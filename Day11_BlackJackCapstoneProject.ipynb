# Functions

```python
import random

cards = [11, 2, 3, 4, 5, 6, 7, 8, 9, 10, 10, 10]


def card_selector(player, cards=cards):
    total = 0
    if len(player) == 0:
        player.append(random.choices(cards, k=2))
        player = player[0]
        return player
    else:
        player.append(random.choice(cards))
        ace_converter(player)
        total = sum(player)
        return player, total


def ace_converter(player):
    if 11 in player:
        if sum(player) > 20:
            idx = player.index(11)
            print("Ace is converted to 1")
            player[idx] = 1
            total = sum(player)
            return player, total


def check_winner(player, dealer):
    global winner
    if sum(player) == 21 and sum(dealer) != 21:
        winner = 'player'
    elif sum(player) > 21 and sum(dealer) > 21:
        winner = 'Game busted'
    elif sum(player) > 21 and sum(dealer) < 21:
        winner = 'Dealer'
    elif sum(player) < 21 and sum(dealer) > 21:
        winner = 'player'
    elif sum(player) < 21 and sum(dealer) < 21 and sum(dealer) < sum(player):
        winner = 'player'
    elif sum(player) < 21 and sum(dealer) < 21 and sum(dealer) > sum(player):
        winner = 'dealer'
    elif sum(player)<21 and sum(dealer)==21:
        winner='dealer'
    elif sum(player) == sum(dealer):
        winner = "Game Level"
    return winner

def play_game(cont,dealer,player):
    print(f'Your cards are {player} and dealers face-up card is {dealer[0]}')
    while cont=='y':
        cont = input("Do you want to pick another card or Pass? type 'y' to pick and 'n' to pass : ")
        if cont=='y':
            player,total_p = card_selector(player)
            if sum(dealer)<17:
                dealer,total_d = card_selector(dealer)
            if sum(player)>21:
                print(f"Your cards are {player} and Your total is {total_p}, Your Game is Busted")
                total_d=sum(dealer)
                break
        elif cont=='n':
            total_p=sum(player)
            total_d=sum(dealer)
        print(f"Your cards are {player} and Your total is {total_p}")
    print(f"Dealer cards are {dealer} and his total is {total_d}")
    winner = check_winner(player,dealer)
    print(f"Winner is {winner}")
```

```python
play = input('Do you want to play the Blackjack game? Type "y" to continue "n" to exit  :').lower()
# print(logo_blackjack)
while play=='y':

    player =[]
    dealer = []
    cont='y'

    player = card_selector(player)
    dealer = card_selector(dealer)

    play_game(cont=cont,dealer=dealer,player=player)
    play = input('Do you want to play another game ? Type y to continue or n to exit : ')
```
