# If else elif in Roller Coaster ride

```python
height = int(input('What is your height in cms? '))
age = int(input('What is your age buddy? '))

if height <120:
  print('Sorry you cant ride the roller coaster ride')
elif height>=120 and age<18:
  print('Your ticket price is $7')
else:
  print('Your ticket price is $12')
```

# Nested if else condition

```python
height = int(input('What is your height in cms? '))
age = int(input('What is your age buddy? '))

if height >=120:
  if age>=18:
    print('Your ticket price is 12$')
  else: 
    print('Your ticket price is $7')
else:
  print('Sorry you cant ride the roller coaster ride')
```

# Leap Year Determination

image.png

```python
# 🚨 Don't change the code below 👇
year = int(input("Which year do you want to check? "))
# 🚨 Don't change the code above 👆

if year%4==0:
  if year%100==0:
    if year%400==0:
      print('Leap year.')
    else:
      print('Not leap year.')
  else:
    print('Leap year.')
else:
  print('Not leap year.')
```

# Multiple If Statement in Succession

```python
height = int(input('What is your height in cms? '))
age = int(input('What is your age buddy? '))

if height >=120:
  if age<12:
    bill = 5
    print(f'Your Bill amount would be {bill}')
  elif age<18:
    bill = 7
    print(f'Your Bill amount would be {bill}')
  else:
    bill = 12
    print(f'Your Bill amount would be {bill}')
  pic = input('Do you need a photo? Provide Y or N')
  if pic=="Y":
    bill+=3
    print(f'Your Total bill amount is {bill}')
  else:
    print(f'Your Total bill amount is {bill}')
else:
  print('Sorry you cant ride the roller coaster ride, your height is too short')
```

# Python Pizza

```python
# 🚨 Don't change the code below 👇
print("Welcome to Python Pizza Deliveries!")
size = input("What size pizza do you want? S, M, or L ")
add_pepperoni = input("Do you want pepperoni? Y or N ")
extra_cheese = input("Do you want extra cheese? Y or N ")
# 🚨 Don't change the code above 👆

#Write your code below this line 👇
bill = 0
if add_pepperoni=='Y' and size=='S':
    bill+=2
elif add_pepperoni=='Y' and (size =='L' or 'M'):
    bill+=3

if extra_cheese=='Y':
    bill+=1

if size=="S":
    bill+=15
elif size =='M':
    bill+=20
elif size=='L':
    bill+=25

print(f'Your final bill is: ${bill}.')
```

```python
# 🚨 Don't change the code below 👇
print("Welcome to the Love Calculator!")
name1 = input("What is your name? \n")
name2 = input("What is their name? \n")
# 🚨 Don't change the code above 👆

#Write your code below this line 👇
name1 = name1.lower()
name2 = name2.lower()
name = name1+name2

T= name.count('t')
R= name.count('r')
U= name.count('u')
E= name.count('e')
true_word= T+R+U+E
true_word = str(true_word)
L=name.count('l')
O=name.count('o')
V=name.count('v')
E=name.count('e')
love_word= L+O+V+E
love_word = str(love_word)
true_love = true_word+love_word
true_love = int(true_love)

if true_love<10 or true_love>90:
    print(f'Your score is {true_love}, you go together like coke and mentos.')
elif true_love>39 and true_love<50:
    print(f'Your score is {true_love}, you are alright together.')
else:
    print(f"Your score is {true_love}.")


```
