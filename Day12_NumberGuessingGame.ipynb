```python
def difficulty_selector():
    '''function to allow user to select the difficulty level and return the number of attempts available'''
    attempts = input("Choose a difficulty level. Type 'easy' or 'hard' : ").lower()
    if attempts == 'easy':
        attempts_number = 10
        return attempts_number
    elif attempts == 'hard':
        attempts_number = 5
        return attempts_number
    else:
        attempts_number = 0
        return attempts_number


def clear():
    from os import system, name
    # for windows the name is 'nt'
    if name == 'nt':
        _ = system('cls')
    else:
        _ = system('clear')





def game_checker(level, required_number):
    won=False
    guess_number = int(input("Make a guess between 0-100 : "))
    level -= 1
    if guess_number == required_number:
        won = True
        print("You Won !")
    return won, level, guess_number
```

```python
import random




game = 'y'
while game == 'y':
    clear()  # Vicks ki goli lo, khich khich door karo, clear screen first
    print("Welcome to the number guessing game!")
    # Select the difficulty level
    level = difficulty_selector()
    # if invalid input it will ask again
    while level == 0:
        print("Invalid input, Type a valid input")
        level = difficulty_selector()

    print(f"You have {level} attempts remaining to make a guess")
    required_number = random.randint(0, 100)
    guess_number = int(input("Make a guess between 0-100 : "))

    print(f"Psst, the solution is {required_number}")  # testing the solution using tester
    won = False
    while level != 1 and won == False:
        print(level)
        if level == 2:
            print('Boss, this is your last attempt. Warna Thakur tu toh Giyo')
        if guess_number == required_number:
            won = True
            print("You Won !")
            break
        elif guess_number > required_number:
            print("Too High, Try again")
            won, level,guess_number = game_checker(level=level, required_number=required_number)
        elif guess_number < required_number:
            print('Too Low, Try Again')
            won, level,guess_number = game_checker(level=level, required_number=required_number)

    game = input(
        f"Game Over ! Correct Answer was {required_number} Do you want to try again? Type \"y\" to play another game, "
        f"\"n\" to exit : ")
```
